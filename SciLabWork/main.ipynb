```scilab
h = 0.05;
N = 1024;
t = 0:h:(N - 1) * h;
```

```scilab
y = (sin(t) + 0.75 * sin(5 / %pi * t)) * 100;
plot2d(t, y)
```

```scilab
Y = fft(y);
PY = abs(Y) .^2 / N;
w = 2 * %pi / h;
dw = w / N;
wd = 0:dw:5;
L = length(wd);
figure(2);
plot2d(wd, PY(1:L))
```

```scilab
y_noize = rand(1:N, 'uniform') * 100 + y;
Y_noize = fft(y_noize, -1);
histplot(10, y_noize);
```

```scilab
Y_noize_magn = abs(Y_noize) .^2 / N;
w = 2 * %pi / h;
dw = w / N;
wd = 0:dw:5;
L = length(wd);
figure(2);
plot2d(wd, Y_noize_magn(1:L))
```
