```python
import cv2 as cv
from matplotlib import pyplot as plt
import numpy as np
```

```python
def show(img, ax=None):
    if ax == None:
        plt.imshow(img[:, :, ::-1])
    else:
        ax.imshow(img[:, :, ::-1])
```

```python
def image_info(img):
    shape = img.shape
    print(f'Height: {shape[0]}')
    print(f'Width: {shape[1]}')
    print(f'Channels: {shape[2]}')
    print(f'Dtype: {img.dtype}')
    print(f'{img.min() = }, {img.max() = }')
```

```python
img1 = cv.imread(r'Gazprom.jpg')
img2 = cv.imread(r'RN.jpg')
```

```python
image_info(img1)
```

```python
image_info(img2)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
show(img1, ax1)
show(img2, ax2)
```

```python
img1_add = img1 + 1
show(img1_add)
```

```python
img1_resized = cv.resize(img1, (256, 256), cv.INTER_CUBIC)
show(img1_resized)
```

```python
image_info(img1_resized)
```

```python
gamma = 0.9
img1_pow = (img1 ** gamma).astype(np.uint8)
show(img1_pow)
```

```python
img2_f64 = (img2 / 255).astype(np.float64)
show(img2_f64)
```

```python
def apply_conv(img_part, w, size):
    return (img_part * w).sum()
```

```python
def apply_filter(img, w):
    G = np.zeros_like(img)
    size = w.shape[0]

    for i in range(img.shape[0] - size):
        for j in range(img.shape[1] - size):
            G[i, j] = apply_conv(img[i:i+size, j:j+size], w, size)
    return G
```

```python
w = np.array(
    [[-1, -1, -1], 
    [ 0,  0,  0],
    [ 1,  1,  1]]
) / 9
```

```python
%%time
G = apply_filter(img1, w)
```

```python
show(G)
```
